# Motor Trends Cars

Explore the cars dataset.

```python
"""Motor Trends Cars Dataset."""

from pathlib import Path

import matplotlib.pyplot as plt
import polars as pl

import config

p = (
    Path(config.PROCESSED_DATA_DIR)
    / "cars"
    / "mtcars.parquet"
)
df = pl.read_parquet(p)
```

Generate a graph.

```python
# Histogram
fig, ax = plt.subplots(figsize=(6, 4))
ax.hist(df.get_column("mpg"), bins=10, color="#4C72B0", edgecolor="white")
ax.set_title("mpg distribution")
ax.set_xlabel("mpg")
ax.set_ylabel("count")
fig.tight_layout()
plt.show()
```

Save the plot to the reports/figures directory.

```python
output_path = Path(config.FIGURES_DIR) / "cars_mpg.png"
fig.canvas.draw()
fig.savefig(output_path, dpi=300, bbox_inches="tight")
print(f"Saved figure to {output_path}")
```
